```python
#finally, loading in machine learning data and addind stage data for patients
machine_learning_dataset = pd.read_csv('~/Desktop/Main/School_Files/S2/QBIO/qbio_490_alexzhang/FInal/TRUE_FINAL_MACHINE_LEARNING_DF.csv')
machine_learning_dataset.rename(columns={'Matched_Norm_Sample_Barcode': 'patient'}, inplace = True)
machine_learning_dataset.set_index('patient', inplace = True)

#load rna_clinical
rna_clinical = pd.read_csv('~/Desktop/Main/School_Files/S2/QBIO/qbio_490_alexzhang/FInal/rna_clinical.csv', index_col='patient')
rna_clinical_intermed = rna_clinical[['ajcc_pathologic_stage']].copy()
rna_clinical_intermed.reset_index(inplace=True)
rna_clinical_intermed.rename(columns={'index': 'patient'}, inplace = True)
rna_clinical_intermed.set_index('patient', inplace = True)

#merge
ml_dataframe_final = pd.merge(machine_learning_dataset, rna_clinical_intermed, on = 'patient', how = 'left')
ml_dataframe_COMPLETE = ml_dataframe_final.drop_duplicates()

#changing [Stage X] to [X]
mapping_dict = {"Stage III": 3, "Stage II": 2, "Stage I": 1, "Stage IV": 4}
ml_dataframe_COMPLETE["ajcc_pathologic_stage"] = ml_dataframe_COMPLETE["ajcc_pathologic_stage"].replace(mapping_dict)
ml_FR_final = ml_dataframe_COMPLETE.fillna(0)
```

```python
##MACHINE LEARNING, 80% train 20% test split##
#import
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#machine learning code
test_dataset = ml_FR_final
scaler = StandardScaler()
data = scaler.fit_transform(ml_FR_final.to_numpy())

classifier = [KNeighborsClassifier(), DecisionTreeClassifier(), MLPClassifier(max_iter=500), GaussianNB()]

classifier_names = ['Nearest Neighbors', 'Decision Tree', 'Multilayer Perceptron', 'Naive-Bayes']

fig, axs = plt.subplots(1, 4, constrained_layout=True, sharey=True)

colors = ['red', 'blue', 'green', 'purple']

for split in np.arange(0.05, 1, 0.05):
    X_train, X_test, y_train, y_test = train_test_split(data, test_dataset.ajcc_pathologic_stage, train_size=0.8)

    for i in range(len(classifier)):
        classifier[i].fit(X_train, y_train)
        y_pred = classifier[i].predict(X_test)
        fraction_correct = sum(y_pred==y_test)/len(y_test)
        axs[i].scatter(split, fraction_correct, color=colors[i], alpha=0.5) 

for i in range(4):
    axs[i].set_title(classifier_names[i])
```
